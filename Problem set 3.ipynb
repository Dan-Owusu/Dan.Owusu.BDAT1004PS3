```python
import pandas as pd
import matplotlib.pyplot as pp
mardiv = pd.read_csv('marriage.csv')
mardiv.head(3)
```

```python
mardiv['marperCapita'] = (mardiv['Marriages']/mardiv['Population'])
mardiv['divperCapita'] = (mardiv['Divorces']/mardiv['Population'])
mardiv.head(3)
```

```python
pp.plot(mardiv.loc[:,'Year'], mardiv.loc[:,'marperCapita'], color = 'blue', label = 'Marriage per Capita')
pp.plot(mardiv.loc[:,'Year'], mardiv.loc[:,'divperCapita'], color = 'red', label = 'Divorce per Capita')
pp.title('A line Plot showing the number of marriages and divorces per capita in the US') 
pp.xlabel('Years')
pp.ylabel('marriage/divorce per capita')
pp.legend()
```
